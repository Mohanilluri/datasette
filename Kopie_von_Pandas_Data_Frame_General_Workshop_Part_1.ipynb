## Data Frame Selection / Indexing

```python
students_df = pd.read_csv('students.csv')
students_df
```

### Selecting by Label
First we'll try selecting data by the labels of columns and rows using the <b>.loc[ ]</b>  function.

- In the square brackets we first put the labels of the rows and after the comma the labels of the columns. 

- <b>:</b> instead of a label means that we want to select all of the rows/columns. 

- <b>:</b> added before or after a label means we want to select all columns/rows up to this label or from this label until the end. 

- To select multiple labels we add them as a list - in square brackets, separated by commas. 

```python
#here we want to select all rows, but only columns 'city' and 'age':
students_df.loc[:,["city","age"]]
```

```python
# here we want to select rows from label 0 until label 4 (which is included), and all columns:
students_df.loc[0:4,:]
```

Here are some other ways of selecting one or multiple columns:

```python
students_df[["age","city"]]
```

```python
students_df.city
```

```python
students_df[["city"]]
```

Notice that one Pandas column is a Pandas Series Object: 

```python
type(students_df.city)
```

### Selecting by position / index

To select data by position we'll use the <b>.iloc[ ]</b> function.

- Instead of using the row/column label you use it's index to select data.

- Python indexing is zero-based which means the first row or column can be selected by the index 0, not 1! 

- When you want to use a slice, for example [0:5] the last index will not be selected, the selection goes up to that index!  

```python
# here we are selecting first row and all columns: 

students_df.iloc[0,:]
```

```python
# here we are selecting rows from 1 until 5 as well as the first and the second column:

students_df.iloc[1:6, [0, 1]]
```

To select contents of one cell only you can use .loc[ ] and .iloc[ ] but it's better to use functions <b>.at[ ]</b> (with labels) and <b>.iat[ ]</b> (with index).

```python
students_df.at[1, 'name']
```

```python
students_df.iat[2, 0]
```

## Setting / Updating data

### Updating index in the DataFrame

Let us first update the DataFrame index with the <b>.index</b> function.

```python
# assign new indices using a list: 

students_df.index = ["a","b","c","d","e","f","g"]
students_df
```

```python
# with the same function we can check the current indices 

students_df.index
```

### Updating values in the DataFrame by constant

It's possible to select data using the .loc and .iloc functions and then assign a constant value to it like this: 

```python
students_df.loc[:'e', 'py-score'] = 70
students_df
```

### Updating values using a list of values 
 
You can assign a list of values to selected rows/columns.
<br><br>
Notice how to access the last item!

```python
students_df.iloc[:, -1] = [88.0, 90, 81.0, 80.0, 68.0, 61.0, 84.0]
students_df
```

### Updating by string value also works!

```python
students_df["py-score"]="static value"
students_df
```

```python
## reload the data to undo changes and update index: 

students_df = pd.read_csv('students.csv')
students_df.index = ["a","b","c","d","e","f","g"]
students_df
```

### Updating a column using a formula

You can update all values in a column using a mathematical formula.

```python
# here we'll multipy all values in column "py-score" by 1.5 

students_df["py-score"]=students_df["py-score"]  * 1.5
students_df
```

Here is another way how to do it:

```python
students_df["py-score"] +=  10 
students_df
```

And a more advanced way using the <b>map</b> function: 

```python
students_df["py-score"] = list(map(lambda x: x+10,students_df["py-score"]))
students_df
```

# Inserting / deleting rows 

You can insert a row into a DataFrame using the <b>.append()</b> function and delete a row using the <b>.drop()</b> function.

```python
# to append a row use a dictionary where keys equal column names and values equal the contents of each cell:
 
students_df=students_df.append({'name':"Marcus", 'city':"Berlin","age":35,"py-score":75}, ignore_index=True)
students_df
```

```python
# to drop a row use labels of the rows to drop, inplace=True means that the row will be deleted permanently: 

students_df.drop(labels=[7],inplace=True)
students_df
```

### Inserting /deleting columns

To create a new column in the last position of your DataFrame you can use a new column name in square brackets like you were selecting an existing column and assign a list of values to it:

```python
# here we will create a new column labeled 'js-score':

students_df['js-score'] = [71.0, 95.0, 88.0, 79.0, 91.0, 91.0, 80.0]
students_df
```

### Inserting a column into a specific location

To insert a new column into a specific location in your DataFrame you have to use the <b>.insert()</b> function:

```python
# here we will create a new column labeled 'django-score' and locate it on the fifth position:

students_df.insert(loc=4, column='django-score', value=[70, 74, 78, 56, 66, 78, 81.0])
students_df
```

### Inserting a new column using other columns

To create a new column we can also combine the values of other columns using mathematical operators and even add a constant to it at the same time. 

```python
students_df["js-score-updated"] = students_df["js-score"] + students_df["py-score"] + 5
students_df
```

### Deleting a specific column

To drop a column we'll use the <b>.drop()</b> funtion. You select the column to drop by its label. 
<br>
- axis = 0 - dropping by row
- axis = 1 - dropping by column

## Filtering/Boolean Indexing

In Pandas you can select data using a filter / boolean indexing. It means that only if a specific condition is met, the data will be selected.

```python
# here we are selecting only the rows in which the value in the 'py-score' column is larger or equal 140.

students_df[students_df["py-score"] >= 140]
```

```python
# this way we can see in which row the value in the 'age' column is higher than 40

students_df['age'] > 40
```

```python
# here we are not only giving the condition but we are filtering the whole DataFrame to show the rows where the condition is met:

students_df[students_df['age'] > 40]
```

```python
# here we are creating a filter which we can later use to filter the DataFrame

very_good_students_filter = students_df['py-score'] >= 50
very_good_students_filter
```

```python
students_df[very_good_students_filter]
```

### Creating powerful filters with logical operators AND, OR, NOT, XOR

You can create filters combining two or more conditions using logical operators. Then data will be selected only if all conditions are met.

```python
# here we want to select students whose 'py-score' is higher or equal 140 as well as whose 'js-score' is higher or equal 80:

students_df[(students_df['py-score'] >= 140) & (students_df['js-score'] >= 80)]
```

# (Optional) Working with missing data

In a real life scenario we would often encounter missing values in our data. 
<br>
<br>
Missing values in data can be caused by different factors, for example if data is collected through a survey and answering some questions is optional this will result in missing values where the respondent decides not to give an answer.
<br> 
<br> 
There is multiple ways how to deal with missing values - you can drop whole rows that contain missing values, or fill them with other values. What you do would very much depend on the case.  

### NaN or <b>np.nan</b> is used to represent missing values

NaN stands for <b>Not a Number</b>. It is a special floating-point value and cannot be converted to any other type than float.

```python
# first let's create a small DataFrame which will contain NaNs: 

df_ = pd.DataFrame({'x': [1, 2, np.nan, 4]})
df_
```

```python
# Let's add another column: 

df_["y"]=[2,np.nan,4,5]
df_
```

### Dropping rows that contains missing values 

One method o deal with NaNs is to drop whole rows which contain missing values, this however results in loosing a lot of data. 

```python
df_.dropna()
```

Notice that this won't affect the original DataFrame:

```python
df_
```

### Filling missing values with average

You can fill the missing values using the average of each column.

```python
# here we'll fill all NaNs in column x with the average of column x values: 

df_.x.fillna(value=np.mean(df_.x),inplace=True) #using inplace=True will affect the original DataFrame 
df_
```

### Filling missing values with .fillna() function

You can fill the missing values with the .fillna() function - this function offers different methods of filling missing values. 

```python
df_ = pd.DataFrame({'x': [1, 2, np.nan, np.nan]})
print('df_.fillna(value=0)\n',df_.fillna(value=0))
print("\ndf_.fillna(method=ffill)\n",df_.fillna(method='ffill'))
print("\ndf_.fillna(method=bfill)\n",df_.fillna(method='bfill'))
```
